```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding="unicode_escape")
```

```python
#basically shows the number of columns and rows respectively
df.shape
```

```python
#describes the columns and their data types
df.info()
```

```python
#drop a column
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum() #counts the number of null value in each column and displays the total sum
```

```python
#drop null value
df.dropna(inplace=True) # "inplace=True" makes the actual changes in the file
```

```python
pd.isnull(df).sum() # ==> ".sum()" counts the total no. of null value
```

```python
#check the data type of column/attribute
df['Amount'].dtypes
```

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
#check the changed data type of  Amount
df['Amount'].dtypes
```

```python
#check columns
df.columns
```

```python
#.describe() method returns description of the data in the Dataform (i.e., count ,mean ,std ,etc)
df.describe()
```

```python
df.columns
```

# <span style="font-family:Comic Sans MS;"><b>Gender Count</b></span>

```python
#the following loop is used to add the value over the bar
gen=sns.countplot(x='Gender',data=df,hue='Gender',palette='muted',legend=False)
for bars in gen.containers:
    gen.bar_label(bars)
```

# <span style="font-family:Comic Sans MS;"><b>Highest Amount Purchased By Each Gender</b></span>

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Gender',palette='bright',legend=False)
plt.title('Highest Amount Purchased By Each Gender')
plt.xlabel=('GENDER')
plt.ylabel=('AMOUNT')
```

## <span style="font-family:Comic Sans MS;">From the above graphs, we can see that more buyers are female, suggesting that the purchasing power of **female** customers is stronger than that of **male** customers.</span>


# <span style="font-family:Comic Sans MS;"><b>Number of Orders by Age Count</b></span>

```python
gen=sns.countplot(data=df,x='Age Group',hue='Gender',palette='bright',legend=False)
for bars in gen.containers:
    gen.bar_label(bars)
```

# <span style="font-family:Comic Sans MS;"><b>Total Amount Spent by Age Group</b></span>

```python
#total amount Vs Age Group
sns.set(rc={'figure.figsize':(6,5)})
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount',hue='Age Group',palette='bright',legend=False)
plt.title('Total Amount Spent by Age Group',fontweight='bold')
```

## <span style="font-family:Comic Sans MS;"><b>From the above graphs we can say that _most buyers_ are from **26-35** age group and they are **Females**</b></span>

```python
df.columns
```

# <span style="font-family:Comic Sans MS;"><b>Total Number Of Orders From Each State</b></span>

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={"figure.figsize": (25,15)})
sns.barplot(data=sales_state,x='State',y='Orders',hue='State',palette='bright',legend=False)
plt.title('Total number of Orders from each State',fontsize=14,fontweight='bold')
```

## <span style="font-family:Comic Sans MS;"><b>According to the above represented graph, **Uttar Pradesh** and **Maharashtra** are the two top states and followed by the **karnataka** state with the highest **orders!**</b></span>

# <span style="font-family:Comic Sans MS;"><b>Highest Amount Spent By Each State</b></span>

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={"figure.figsize": (25,15)})
sns.barplot(data=sales_state,x='State',y='Amount',hue='State',palette='bright',legend=False)
plt.title('Highest amount spent by each State',fontsize=14,fontweight='bold')
```

## **=>** <span style="font-family:Comic Sans MS;"><b>So basically the data in the above graphs shows up is same as the before, here also **Uttar Pradesh, Mahrashtra** and **Karnataka** are the top states which spent high amount in the sales.</b></span>

## **=>** **<u>Note:</u>**<span style="font-family:Comic Sans MS;"><b> The number of **Orders** and number of **Amount** spent are different by each state as per the given graphs and the analysed data.</b></span>

# <span style="font-family:Comic Sans MS;"><b>Marital Status</b></span>

```python
marital=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status',palette='muted',legend=False)
sns.set(rc={'figure.figsize':(7,5)})
for bars in marital.containers:
    marital.bar_label(bars)
```

## <span style="font-family:Comic Sans MS;"><b>The above represents that <b>married people's</b> count is <b>Higher</b>,so most of the order are from <b>married people!</b></span>

# <span style="font-family:Comic Sans MS;"><b>Amount Spent By Marital Status and By Gender</b></span>

```python
marital_sales=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=marital_sales,hue='Gender',palette='bright',legend=False)
```

## <span style="font-family:Comic Sans MS;"><b>The above Graph shows that _married(Women)_ have a strong purchasing power!</b></span>

# <span style="font-family:Comic Sans MS;"><b>Occupation</b></span>

```python
sns.set(rc={'figure.figsize':(20,5)})
ocup=sns.countplot(x='Occupation',data=df,hue='Occupation',palette='bright',legend=False)

for bars in ocup.containers:
    ocup.bar_label(bars)
```

# <span style="font-family:Comic Sans MS;"><b>Amount Spent By Each Sector In The Sales</b></span>

```python
ocup_sales=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=ocup_sales,x='Occupation',y='Amount',hue='Occupation',palette='bright',legend=False)
```

## **=>** <span style="font-family:Comic Sans MS;">**According to given graph it can be seen that the top Sectors are IT,Healthcare and Aviation who spent most amount in the Sales**</span>

# <span style="font-family:Comic Sans MS;">**Product Category**</span>

```python
sns.set(rc={'figure.figsize':(25,5)})
prod= sns.countplot(data = df, x = 'Product_Category',hue='Product_Category',palette='bright',legend=False)

for bars in prod.containers:
    prod.bar_label(bars)
```

```python
prod_sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = prod_sales, x = 'Product_Category',y= 'Amount',hue='Product_Category',palette='dark',legend=False)
```

## <span style="font-family:'Comic Sans MS';"><b>From the above two graph we can say that most of the sales are from Food,Clothing & Apparel and Electronic Products</b></span>

# <span style="font-family:'Comic Sans MS';"><b>Top Selling Products(with ID)</b></span>

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',hue='Product_ID',palette='bright',legend=False)
plt.title('Top Products with most Orders',fontweight='bold',fontsize=14)
```
